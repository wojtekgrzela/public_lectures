```python
import pandas as pd
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
from termcolor import colored

size = 10000

# Data_Normal:
loc = 11
scale = 3
data_normal1 = np.random.normal(loc, scale, size)

loc = 9
scale = 5
data_normal2 = np.random.normal(loc, scale, size)

loc = 7
scale = 7
data_normal3 = np.random.normal(loc, scale, size)

loc = 5
scale = 9
data_normal4 = np.random.normal(loc, scale, size)

loc = 3
scale = 11
data_normal5 = np.random.normal(loc, scale, size)

'''
# Data_Poisson:
lam = 0.6
data_poisson = np.random.poisson(lam, size)

# Data_Beta
a = 0.4
b = 0.6
data_beta = np.random.beta(a, b, size)

# Data_Exponential
scale = 0.4
data_exponential = np.random.exponential(scale,size)
'''

dict = {'Normal1':data_normal1,
       'Normal2':data_normal2,
       'Normal3':data_normal3,
       'Normal4':data_normal4,
       'Normal5':data_normal5}

data = pd.DataFrame(dict)
# print(data)
# plt.figure()
# data.plot()
data_melted = data.melt()
# print(data_melted)

print(colored("The distribution of variables  is quite narrow. There are quite a lot of bins so it shows how little are the values spreaded.", 'red'))
plt.figure()
grid = sn.FacetGrid(data_melted, col="variable", col_wrap=3)
grid.map(plt.hist, "value",bins=64)


plt.figure()
grid = sn.FacetGrid(data_melted, col="variable", col_wrap=3)
grid.map(plt.hist, "value", cumulative=True,bins=64)
```

```python
from scipy.special import expit

Data_expit = expit(data)
Data_expit_melted = Data_expit.melt()

plt.figure()
grid = sn.FacetGrid(Data_expit_melted, col="variable", col_wrap=3)
grid.map(plt.hist, "value",bins=64)

print(colored("As the distribution was quite narrow then the inversed data is also quite narrow and spikes in the conrners. With the cumulated data there is also a spike in the conrner near value 1.0.", 'red'))

plt.figure()
grid = sn.FacetGrid(Data_expit_melted, col="variable", col_wrap=3)
grid.map(plt.hist, "value", cumulative=True,bins=64)
```

```python
size = 10000

# Data_Poisson:
lam = 1
data_poisson1 = np.random.poisson(lam, size)

lam = 3
data_poisson3 = np.random.poisson(lam, size)

lam = 6
data_poisson6 = np.random.poisson(lam, size)

lam = 9
data_poisson9 = np.random.poisson(lam, size)

lam = 12
data_poisson12 = np.random.poisson(lam, size)


dict = {'Poisson1':data_poisson1,
       'Poisson3':data_poisson3,
       'Poisson6':data_poisson6,
       'Poisson9':data_poisson9,
       'Poisson12':data_poisson12}

data = pd.DataFrame(dict)

data_melted = data.melt()

plt.figure()
grid = sn.FacetGrid(data_melted, col="variable", col_wrap=3)
grid.map(plt.hist, "value",bins=64)

print(colored("Poisson distribution has a standard deviation (sqrt(lambda)). (lambda) is the expected value. It can be seen that the increase of the lambda increases the spread of the calues.", 'red'))

plt.figure()
grid = sn.FacetGrid(data_melted, col="variable", col_wrap=3)
grid.map(plt.hist, "value", cumulative=True,bins=64)
```

```python
size = 10000

# Data_Beta
a = 1
b = 1
data_beta1 = np.random.beta(a, b, size)

a = 5
b = 5
data_beta2 = np.random.beta(a, b, size)

a = 1
b = 5
data_beta3 = np.random.beta(a, b, size)

a = 5
b = 1
data_beta4 = np.random.beta(a, b, size)

a = 0.4
b = 0.6
data_beta5 = np.random.beta(a, b, size)


dict = {'Beta1':data_beta1,
       'Beta2':data_beta2,
       'Beta3':data_beta3,
       'Beta4':data_beta4,
       'Beta5':data_beta5}

data = pd.DataFrame(dict)

data_melted = data.melt()

plt.figure()
grid = sn.FacetGrid(data_melted, col="variable", col_wrap=3)
grid.map(plt.hist, "value",bins=64)

print(colored("Interesting fact is that when (alfa) and (beta) is set to '1' when the probabilities are to the power of 0 thus resulting in an even spread across all values. For other values but equal for alfa and beta the distribution looks a bit like a normal one. When the values are quite fifferent from each othe then the distribution gets shifted th one og the corners.", 'red'))

plt.figure()
grid = sn.FacetGrid(data_melted, col="variable", col_wrap=3)
grid.map(plt.hist, "value", cumulative=True,bins=64)
```

```python
size = 10000

# Data_Lognormal
mean = 1
sigma = 0.1
data_lognormal1 = np.random.lognormal(mean, sigma, size)

mean = 8
sigma = 0.3
data_lognormal2 = np.random.lognormal(mean, sigma, size)

mean = 0
sigma = 0.6
data_lognormal3 = np.random.lognormal(mean, sigma, size)

mean = 1
sigma = 0.9
data_lognormal4 = np.random.lognormal(mean, sigma, size)

mean = 2
sigma = 1
data_lognormal5 = np.random.lognormal(mean, sigma, size)


dict = {'Data_Lognormal1':data_lognormal1,
       'Data_Lognormal2':data_lognormal2,
       'Data_Lognormal3':data_lognormal3,
       'Data_Lognormal4':data_lognormal4,
       'Data_Lognormal5':data_lognormal5}

data = pd.DataFrame(dict)

data_melted = data.melt()

plt.figure()
grid = sn.FacetGrid(data_melted, col="variable", sharex=False, sharey=False, col_wrap=3)
grid.map(plt.hist, "value",bins=64)

print(colored("The larger the sigma gets the narrower the distribution is. Both the mean value and variables values are taken into calculations.", 'red'))

plt.figure()
grid = sn.FacetGrid(data_melted, col="variable", sharex=False, sharey=False, col_wrap=3)
grid.map(plt.hist, "value", cumulative=True,bins=64)
```
