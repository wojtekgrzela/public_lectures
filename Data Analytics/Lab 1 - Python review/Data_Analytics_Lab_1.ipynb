```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from termcolor import colored

data = pd.read_csv('Data1.csv')

data.rename(columns={'Unnamed: 0':'Date'}, inplace=True)
kolumny = data.columns
print(kolumny)


plt.figure()
data = data.set_index('Date')
print(colored("Setting the date as the index so we can reference each row of data by date", 'red'))
print(data)
data.plot()


dataMelted = data.melt()
print(colored("Melting creates an array of columns with variable and value", 'red'))
print(dataMelted)

plt.figure()
grid = sns.FacetGrid(dataMelted, col="variable", col_wrap=3)
grid.map(plt.hist, "value")

# plt.figure()
# sns.distplot(data['theta_1'],kde=True)


for i in kolumny:
    if i == kolumny[0]:
        continue
    plt.figure()
    sns.distplot(data[i], kde=True, hist=False)

print(colored("Melted data allows for values to be assigned to variables. That allows for creating a histogram of number of total occurrences of values.\n", 'red'))
print(colored("It can be seen that in theta_4 and theta_5 that there is inconsistency in the distribution of the values. They should be rescaled to fit into probability density distribution model.", 'red'))



```

```python
'''Data from year: 2018'''

data2018 = data['2018-01-01':'2019-01-01']
kolumny2018 = data.columns[0:4]
data2018 = data2018[kolumny2018]
print(colored("Taking data only from 2018.", 'red'))
print(data2018)
plt.figure()
data2018.plot()
print(colored("Columns from 2018 0-4", 'red'))
print(kolumny2018)

dataMelted2018 = data2018.melt()

plt.figure()
grid = sns.FacetGrid(dataMelted2018, col="variable", col_wrap=2)
grid.map(plt.hist, "value")

# for i in range(0,4):
#     plt.figure()
#     sns.distplot(dataMelted2018[kolumny2018[i]], kde=True, hist=False)

for i in kolumny2018:
    plt.figure()
    sns.distplot(data2018[i], kde=True, hist=False)

print(colored("Interesting fact that can be observed is that the data in one year is quite consistent - there is only one peak in the distribution instead of more in the previous charts where all the data was concerned.", 'red'))
```
